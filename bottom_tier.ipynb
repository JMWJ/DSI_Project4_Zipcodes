```python
import pandas as pd
import numpy as np
```

```python
b_tier = pd.read_csv('./bottom_tier.csv', encoding = "ISO-8859-1")
b_tier.head()
```

```python
cols = ['RegionName', 'State', '2018-11']
b_tier = b_tier[cols]
b_tier.head()
```

```python
b_tier.columns = ['zip', 'state', '2018-11']
```

```python
b_tier.head()
```

```python
b_tier.zip = b_tier.zip.map(lambda cell: str(cell))
```

```python
b_tier.zip.dtypes
```

```python
b_tier.zip = b_tier.zip.map(lambda cell: '0' + cell if len(cell) < 5 else cell)
```

```python
b_tier.drop(columns='zip_length', inplace = True)
```

```python
b_tier.head()
```

```python
b_tier.to_csv('cleaned_b_tier.csv', encoding='utf-8', index=False)
```
